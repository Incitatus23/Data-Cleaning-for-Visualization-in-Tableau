```python
New_England = ["CONNECTICUT", "MAINE", 'MASSACHUSETTS',"NEW_HAMPSHIRE",'RHODE_ISLAND', 'VERMONT']
Mideast = ["DELAWRE", "", 'MASSACHUSETTS',"NEW_HAMPSHIRE",'DISTRICT_OF_COLUMBIA', 'MARYLAND', "NEW JERSEY",
           "NEW YORK", "PENNSLYVANIA"]
Great_Lakes = ["ILLINOIS", "INDIANA", 'MICHIGAN',"OHIO",'WISCONSIN']

Plains = ["IOWA", "KANSAS", 'MINNESOTA',"WISCONSIN",'NEBRASKA', 'NORTH_DAKOTA', "SOUTH_DAKOTA"]

Southeast = ["ALABAMA", "ARKANSAS", 'FLORIDA',"GEORGIA",'KENTUCKY', 'LOUISIANA', "MISSISSIPPI", 
             "NORTH_CAROLINA","SOUTH_CAROLINA", "TENNESSEE", "VIRGINIA", "WEST_VIRGINIA"]

Southwest = ["ARIZONA", "NEW_MEXICO", 'OKLAHOMA',"TEXAS"]

Rocky_Mountains = ["COLORADO", "IDAHO", 'MONTANA',"UTAH",'WYOMING']

Far_West = ["ALASKA", "CALIFORNIA", 'HAWAII',"NEVADA",'OREGON', "WASHINGTON"]
```

```python
education["Region"] = "New England" if education.loc[education["STATE"].isin(New_England)
```

```python
education3 = education.loc[education["ENROLL"].notna()]

education3
```

```python
year_list = [2000, 2001, 2002, 2003,
       2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014,
       2015, 2016]

education4 = education3.loc[education["YEAR"].isin(year_list)]
```

```python
education4 = education4.loc[education["AVG_READING_4_SCORE"].notna()]
```

```python
education4 = education4.loc[education["AVG_MATH_4_SCORE"].notna()]
```

```python
education4 = education4.loc[education["AVG_READING_8_SCORE"].notna()]
```

```python
education4 = education4.loc[education["AVG_MATH_8_SCORE"].notna()]
```

```python
no_states = ["ALASKA", "HAWAII", "DISTRICT_OF_COLUMBIA"]

education5 = education4.loc[~education4["STATE"].isin(no_states)]
```

```python
education5["STATE"].value_counts()
```

```python
education5 = education5[["STATE", "YEAR", "ENROLL", "TOTAL_EXPENDITURE", "INSTRUCTION_EXPENDITURE", "AVG_MATH_4_SCORE",
                       "AVG_MATH_8_SCORE", "AVG_READING_4_SCORE", "AVG_READING_8_SCORE"]]
```

```python
education5
```

```python
education5["EXPENDITURE_PER_PUPIL"] = (education5["INSTRUCTION_EXPENDITURE"] / education5["ENROLL"]).round(2)
```

```python
education5["TOTAL_EXPENDITURE_PER_PUPIL"] = (education5["TOTAL_EXPENDITURE"] / education5["ENROLL"]).round(2)
```

```python
education5
```

```python
education5["ENROLL"] = education5["ENROLL"].astype(int)
```

```python
education5
```

```python
education_df = education5.copy()
```

```python
education_df
```

```python
education_df["Previous_Year_Avg_Math_4_Score"] = education_df.groupby("STATE")["AVG_MATH_4_SCORE"].shift(1)
```

```python
education_df["Previous_Year_Avg_Math_8_Score"] = education_df.groupby("STATE")["AVG_MATH_8_SCORE"].shift(1)
```

```python
education_df["Previous_Year_Avg_Reading_4_Score"] = education_df.groupby("STATE")["AVG_READING_4_SCORE"].shift(1)
```

```python
education_df["Previous_Year_Avg_Reading_8_Score"] = education_df.groupby("STATE")["AVG_MATH_8_SCORE"].shift(1)
```

```python
education_df["Previous_Year_Expenditure_Per_Pupul"] = education_df.groupby("STATE")["EXPENDITURE_PER_PUPIL"].shift(1)
```

```python
education_df["Previous_Year_Total_Expenditure_Per_Pupil"] = education_df.groupby("STATE")["TOTAL_EXPENDITURE_PER_PUPIL"].shift(1)
```

```python
education_df2 = education_df.copy()
```

```python
education_df2.to_csv("education_df.csv")
```

```python
education_df2
```

```python
education_df2["MATH_4_RANK_"] = education_df2.groupby("AVG_MATH_4_SCORE")[["YEAR", "STATE"]].rank()
```

```python
education_df2 = education_df2.drop("MATH_4_RANK", axis=1)
```

```python
education_df2
```

```python
education_df2["MTH_8_RANK"] = education_df2.groupby(["YEAR"])["AVG_MATH_8_SCORE"].rank(method="min", ascending=False)
education_df2["READING_4_RANK"] = education_df2.groupby(["YEAR"])["AVG_READING_4_SCORE"].rank(method="min", ascending=False)
education_df2["READING_8_RANK"] = education_df2.groupby(["YEAR"])["AVG_READING_8_SCORE"].rank(method="min", ascending=False)
education_df2["EXPENDITURE_PER_PUPIL_RANK"] = education_df2.groupby(["YEAR"])["EXPENDITURE_PER_PUPIL"].rank(method="min", ascending=False)
education_df2["TOTAL_EXPENDITURE_PER_PUPIL_RANK"] = education_df2.groupby(["YEAR"])["TOTAL_EXPENDITURE_PER_PUPIL"].rank(method="min", ascending=False)
```

```python
education_df2 = education_df2.drop("MTH_4_RANK_2", axis=1)
```
