```python
crime_df["violent_per_100k_rank"] = crime_df.groupby("year")["violent_per_100k"].rank(method="min", ascending=False)
```

```python
crime_df["homs_per_100k_rank"] = crime_df.groupby("year")["homs_per_100k"].rank(method="min", ascending=False)
crime_df["rape_per_100k_rank"] = crime_df.groupby("year")["rape_per_100k"].rank(method="min", ascending=False)
crime_df["rob_per_100k_rank"] = crime_df.groupby("year")["rob_per_100k"].rank(method="min", ascending=False)
crime_df["agg_ass_per_100k_rank"] = crime_df.groupby("year")["agg_ass_per_100k"].rank(method="min", ascending=False)
```

```python
crime_df.head()
```

```python
crime_df["previous_violent_rate"] = crime_df.groupby("department_name")["violent_per_100k"].shift(-1)
```

```python
crime_df.drop("previous_violent_rate", axis=1, inplace=True)
```

```python
crime_df["previous_violent_rate"] = crime_df.groupby("department_name")["violent_per_100k"].shift(1)
```

```python
crime_df["previous_homicide_rate"] = crime_df.groupby("department_name")["homs_per_100k"].shift(1)
```

```python
crime_df["previous_rape_rate"] = crime_df.groupby("department_name")["rape_per_100k"].shift(1)
crime_df["previous_robbery_rate"] = crime_df.groupby("department_name")["rob_per_100k"].shift(1)
crime_df["previous_assault_rate"] = crime_df.groupby("department_name")["agg_ass_per_100k"].shift(1)
```

```python
crime_df["violent_change"] = (crime_df["violent_per_100k"] - crime_df["previous_violent_rate"])/crime_df["previous_violent_rate"]
        
```

```python
crime_df["homicide_change"] = (crime_df["homs_per_100k"] - crime_df["previous_homicide_rate"])/crime_df["previous_homicide_rate"]
crime_df["rape_change"] = (crime_df["rape_per_100k"] - crime_df["previous_rape_rate"])/crime_df["previous_rape_rate"]
crime_df["robbery_change"] = (crime_df["rob_per_100k"] - crime_df["previous_robbery_rate"])/crime_df["previous_robbery_rate"]
crime_df["assault_change"] = (crime_df["agg_ass_per_100k"] - crime_df["previous_assault_rate"])/crime_df["previous_assault_rate"]
        
        
```
